```python
# Import libraries
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
#TASK 1

# ===========================
# Task: Hadamard on Two Qubits
# ===========================
print("\n===== TASK: HADAMARD ON TWO QUBITS =====")

# Step 1: Create a quantum circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)

# Step 2: Apply a Hadamard gate on both qubits
qc.h(0) #or qc.h([0, 1])
qc.h(1)

# Step 3: Measure both qubits
qc.measure([0, 1], [0, 1])

# Draw the circuit
print("\nQuantum Circuit:")
print(qc.draw('text'))

# Step 4: Simulate using AerSimulator
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=1024)
result = job.result()
counts = result.get_counts()

# Step 5: Display the results
print("\nMeasurement Results (Hadamard on Both Qubits):", counts)
display(plot_histogram(counts))

#The histogram shows equal probabilities for all four states, indicating that the
#two-qubit system is in a balanced superposition.
```

```python
#TASK 2

# ===========================
# Task: Swap Qubit Roles (X on second qubit)
# ===========================
print("\n===== TASK: SWAP QUBIT ROLES (X on second qubit) =====")

# Step 1: Create a quantum circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)

# Step 2: Apply X gate on second qubit
qc.x(1)      # Apply X on qubit 1

# Step 3: Measure both qubits
qc.measure([0, 1], [0, 1])

# Draw the circuit
print("\nQuantum Circuit:")
print(qc.draw('text'))

# Step 4: Simulate using AerSimulator
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=1024)
result = job.result()
counts = result.get_counts()

# Step 5: Display the results
print("\nMeasurement Results ( X on second qubit):", counts)
display(plot_histogram(counts))
```

```python
#TASK 3

# ===========================
# Task: Add CNOT Gate (Create Entanglement)
# ===========================
print("\n===== TASK: HADAMARD + CNOT (ENTANGLEMENT) =====")

# Step 1: Create a quantum circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)

# Step 2: Apply Hadamard gate on the first qubit
qc.h(0)

# Step 3: Add CNOT gate (control = qubit 0, target = qubit 1)
qc.cx(0, 1)

# Step 4: Measure both qubits
qc.measure([0, 1], [0, 1])

# Draw the circuit
print("\nQuantum Circuit:")
print(qc.draw('text'))

# Step 5: Simulate using AerSimulator
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=1024)
result = job.result()
counts = result.get_counts()

# Step 6: Display the results
print("\nMeasurement Results (Hadamard + CNOT):", counts)
display(plot_histogram(counts))

#This is a Bell State, also known as |Φ⁺⟩, one of the four maximally entangled states.
```

```python
#TASK 4

# ===========================
# Task: Increase shot count from 1024 to 4096
# ===========================
print("\n===== TASK: INCREASING SHOT COUNT FROM 1024 TO 4096 =====")

# Step 1: Create a quantum circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)

# Step 2: Apply Hadamard gate on the first qubit
qc.h(0)

# Step 3: Add CNOT gate (control = qubit 0, target = qubit 1)
qc.cx(0, 1)

# Step 4: Measure both qubits
qc.measure([0, 1], [0, 1])

# Draw the circuit
print("\nQuantum Circuit:")
print(qc.draw('text'))

# Step 5: Simulate using AerSimulator
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=4096)
result = job.result()
counts = result.get_counts()

# Step 6: Display the results
print("\nMeasurement Results :", counts)
display(plot_histogram(counts))

#when you increase the number of shots from 1024 to 4096,
#the distribution becomes more stable.
#More shots → less noise → more accurate and stable probability distribution.
```

```python
#TASK 5

from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import matplotlib.pyplot as plt
from qiskit.quantum_info import Statevector

# ===========================
# Create a 2-qubit entangled circuit
# ===========================
qc = QuantumCircuit(2, 2)

# Step 1: Apply Hadamard to qubit 0
qc.h(0)

# Step 2: Apply CNOT (control = 0, target = 1)
qc.cx(0, 1)

# ===========================
# Before Measurement — visualize Bloch sphere
# ===========================
state_before = Statevector.from_instruction(qc)
print("\nStatevector before measurement:")
display(plot_bloch_multivector(state_before))

# Step 3: Measure both qubits
qc.measure([0, 1], [0, 1])

# Visualize the circuit
print("\nQuantum Circuit:")
print(qc.draw('text'))

# ===========================
# Simulate measurement
# ===========================
simulator = AerSimulator()
job = simulator.run(qc, backend=simulator, shots=4096)
result = job.result()
counts = result.get_counts()

# Display histogram of results
print("\nMeasurement Results:", counts)
plot_histogram(counts)
plt.show()

# ===========================
# After Measurement — visualize Bloch sphere
# ===========================
# Note: After measurement, qubits collapse to a classical state
collapsed_state = Statevector.from_int(0, 2)  # One possible post-measurement state
print("\nStatevector after measurement (collapsed):")
display(plot_bloch_multivector(collapsed_state))
```
